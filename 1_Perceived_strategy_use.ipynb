#### Functions

```python
def table_porc(D):
    df = D.apply(pd.value_counts).fillna(0).astype(int).T
    total_row = df.sum(axis=1)
    for i in df.index:
        df.loc[i] = df.loc[i]*100/total_row.loc[i]
    df['n'] = total_row
    return df
```

## Analysis

The reported results correspond to the part of the survey study Kienberger (2020) that focuses on specific lexical inferencing strategies' use (part A3). The participants were asked to rate the frequency of the indicated strategies' use on a 5-point Likert scale from "(1) No lo hago nunca o casi nunca" (Never or almost never true of me) to "(5) Lo hago siempre o casi siempre" (Always or amost always true of me). See (Link) and (PDF) for details.
